```python
import pandas as pd
import numpy as np
mixes_1 = pd.read_excel('mixes_2017-2021_b1.xlsx',
     engine='openpyxl')
```

```python
target = mixes_1[['Дата', 'Качество кокса,% (Показатели прочности,% — M25)', 
                  'Качество кокса,% (Показатели прочности,% — M10)', 
                  'Качество кокса,% (Показатели прочности,% — CSR)']]
target.columns = ["ДатаПробыКокса", "M25", "M10", "CSR"]
target["ДатаШихтовки"] = target["ДатаПробыКокса"] - pd.Timedelta(4, "d")
target
```

```python
features = mixes_1[["Дата", "Состав шихты,%(спек.)", "Состав шихты,%(в т.ч. Марка Ж)", "Состав шихты,%(в т.ч. кокс.)",
                    "Оборот печей, час", "Качество шихты (помол,%)", "Качество шихты (пыль,%)", 
                    "Качество шихты (Технический анализ, % — влага)", "Качество шихты (Технический анализ, % — зола)",
                    "Качество шихты (Технический анализ, % — летуч.)", "Качество шихты (Мд. Серы,%)",
                    "Качество шихты (пласт. слой мм)", 'R0', 'σ', 'Vt', 'I', 'Io', 'SI', 'КТЦ эксп.', 'MF', 
                    'MF_spec',  'MF_otosh', 'CSR_carb']]


features.columns = ["ДатаШихтовки", "% спек", "% марка Ж", "% кокс.", "Оборот печей, ч", "Помол, %",
                    "Пыль, %", "Влага, %", "Зола, %", "Летуч., %", "Мд. Серы, %", "Пласт. слой, мм", 
                    'R0', 'σ', 'Vt', 'I', 'Io', 'SI', 'КТЦ эксп.', 'MF', 
                    'MF_spec',  'MF_otosh', 'CSR_carb']

features
```

```python
data_aligned_b1 = pd.merge(target, features)
data_aligned_b1["Оборот печей отсеч, ч"] = np.where(data_aligned_b1['Оборот печей, ч'] >= 18, 18, data_aligned_b1['Оборот печей, ч'])
data_aligned_b1
```
